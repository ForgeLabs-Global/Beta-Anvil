```python
# Loading the Pandas Library
import pandas as pd

# Loading the dataset
df = pd.read_csv('StudentsPerformance (1).csv')

def calculate_math_score_spread(data):

    #Finding the highest score
    highest_score = data['math score'].max()

    #Finding the lowest score
    lowest_score = data['math score'].min()

    #Subtracting values
    spread = highest_score - lowest_score

    return highest_score, lowest_score, spread

# Loading the dataset
df = pd.read_csv('StudentsPerformance (1).csv')

# Using the function
high, low, spread = calculate_math_score_spread(df)

print(f"Highest Math Score: {high}")
print(f"Lowest Math Score: {low}")
print(f"Spread (Difference): {spread}")
```
