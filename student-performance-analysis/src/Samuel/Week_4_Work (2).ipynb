Step 1-3 : Loading the Dataset and Necessary Packages


```python
# Step 1: Loading the Necessary Packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Loading the dataset
df = pd.read_csv('StudentsPerformance.csv')

#Step 3: Creation of countplot based on Test Preparation Course by Parental Level of Education
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='parental level of education', hue='test preparation course')
plt.title('Test Preparation Course by Parental Level of Education')
plt.xticks(rotation=45)
plt.show()
```

Step 4: Addition of the Average and the Total Score Column

```python
# Step 4: We will Add 'Total_score' and 'average' columns

df['Total_score'] = df['math score'] + df['reading score'] + df['writing score']
df['average'] = df['Total_score'] / 3
```

Step 5: Then we Add Math_PassStatus. This is a way that we can implement a cleaning method so know the Pass Mark

```python
#Step 5: Adding 'Math_PassStatus' (Pass mark is 50)

df['Math_PassStatus'] = df['math score'].apply(lambda x: 'P' if x >= 50 else 'F')
```

Step 6: Count Plot for Math_PassStatus

```python
# Count Plot Visualization for the Math_PassStatus
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Math_PassStatus')
plt.title('Math Pass Status Count')
plt.show()
```

Step 7 : Adding  'Reading_PassStatus' (Pass mark is 50)

```python
df['Reading_PassStatus'] = df['reading score'].apply(lambda x: 'P' if x >= 50 else 'F')
```

Step 8:  Adding  'Writing_PassStatus' (Pass mark is 50)

```python
df['Writing_PassStatus'] = df['writing score'].apply(lambda x: 'P' if x >= 50 else 'F')
```

Step 9: Added 'Overall_PassStatus'
 Notw: You will have to  pass Math, Reading, and Writing to get a P

```python
df['Overall_PassStatus'] = df.apply(lambda row: 'P' if (row['Math_PassStatus'] == 'P' and
                                                        row['Reading_PassStatus'] == 'P' and
                                                        row['Writing_PassStatus'] == 'P') else 'F', axis=1)
```

Matthew Tasks- Checking for Empty Values

```python
# Matthew Tasks - Checking for Empty Values (Using the actual dataset)

# 1. Print missing values per column for our working DataFrame (df)
print("Missing values per column:")
missing_vals = df.isnull().sum()
print(missing_vals)

# 2. Count total missing values in the entire dataset
total_missing = missing_vals.sum()
print(f"\nTotal missing values: {total_missing}")

# 3. Standardize all column names (replacing spaces with underscores and making lowercase)
df.columns = df.columns.str.replace(' ', '_').str.replace('/', '_').str.lower()
print("\nCleaned Columns:")
print(df.columns.tolist())

# 4. Export the completely cleaned and updated dataset
df.to_csv("cleaned_data.csv", index=False)
print("\nExport successful! 'cleaned_data.csv' is ready for the Analysis Group.")
```
