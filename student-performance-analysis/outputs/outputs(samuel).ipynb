```python
import pandas as pd

# Load dataset
data = pd.read_csv(r"C:\Users\HP\Documents\GitHub\Beta-Anvil\student-performance-analysis\data\StudentsPerformance.csv")

# =========================
# WEEK 1 — BASICS
# =========================

# Variables
student1 = data["math score"][0]
student2 = data["math score"][1]
student3 = data["math score"][2]

total_score = student1 + student2 + student3

print("WEEK 1 OUTPUT")
print("Total of first 3 students:", total_score)

# Lists + Loop (Pass Count)
math_scores = list(data["math score"])

pass_count = 0
for score in math_scores:
    if score >= 50:
        pass_count += 1

print("Number of students that passed:", pass_count)

# Conditional (First 5 students)
print("\nPass/Fail for first 5 students:")
first_five_scores = math_scores[:5]

for score in first_five_scores:
    if score >= 50:
        print("Pass")
    else:
        print("Fail")


# =========================
# WEEK 2 — FUNCTIONS + FILTERING
# =========================

def average_math_by_gender(gender_value):
    scores = []

    for i in range(len(data)):
        if data["gender"][i] == gender_value:
            scores.append(data["math score"][i])

    return sum(scores) / len(scores)

female_avg = average_math_by_gender("female")
male_avg = average_math_by_gender("male")

print("\nWEEK 2 OUTPUT")
print("Average math score for females:", female_avg)
print("Average math score for males:", male_avg)


# =========================
# WEEK 3 — SCORE SEGMENTATION
# =========================

def separate_scores(scores):
    above_70 = []
    below_or_equal_70 = []

    for score in scores:
        if score > 70:
            above_70.append(score)
        else:
            below_or_equal_70.append(score)

    return above_70, below_or_equal_70


high_scores, low_scores = separate_scores(math_scores)

print("\nWEEK 3 OUTPUT")
print("Number of students above 70:", len(high_scores))
print("Number of students 70 and below:", len(low_scores))

# Optional preview
print("Sample high scores:", high_scores[:5])
print("Sample low scores:", low_scores[:5])
```
