# ⚒ SECTION 2 — CLEANING MODULE

```markdown
# 🧹 Step 1: Create cleaning.py

Inside the modules folder, create:

modules/cleaning.py

This file will handle all cleaning logic.
```

````python
## 👉 In `cleaning.py`

import pandas as pd

def load_data(path):
    return pd.read_csv(path)

def clean_data(df):
    # Hint: remove missing values
    df = df.dropna()
    return df

def save_data(df, path):
    df.to_csv(path, index=False)
````


# 📊 SECTION 3 — ANALYSIS MODULE

```markdown
# 📊 Step 2: Create analysis.py

Inside modules folder:

modules/analysis.py

This file will handle analysis logic.
```

````python
## 👉 In `analysis.py`

def calculate_average(column):
    return column.mean()

def count_pass(column):
    # Hint: count values >= 50
    return (column >= 50).sum()
````

# ⚒ SECTION 4 — MAIN FILE

```markdown
# ⚒ Step 3: Create main.py

This is where everything comes together.
```

````python
## 👉 In `main.py`

from modules.cleaning import load_data, clean_data, save_data
from modules.analysis import calculate_average, count_pass

# Load data
df = load_data("data/students.csv")

# Clean data
df = clean_data(df)

# Save cleaned data
save_data(df, "data/cleaned_data.csv")

# Analysis
math_avg = calculate_average(df["math score"])
pass_count = count_pass(df["math score"])

print("Average Math Score:", math_avg)
print("Students Passed:", pass_count)
````
